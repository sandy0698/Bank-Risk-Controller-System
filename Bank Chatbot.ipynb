```python
llm_model = CTransformers(
    model = model_file,
    model_type = "llama",
    config = {"max_new_tokens":1000, "temperature": 0.01}
)
```

```python
#prompy Engineering
```

```python
"""
You are an AI assistant designed to help users with information about the State Bank Vishwa Yatra Foreign Travel Card.
Below is a list of Frequently Asked Questions (FAQs) about the card. Based on these FAQs, please provide answers to user queries.

Example: Can I use the Add-on cards simultaneously?
bot answer : Yes, Add-on cards can be used simultaneously along with the original card

"""
```

```python
prompt = """

use the following information and answer the question based on that /
provide only useful informations /
if you do not know, do not try to make wrong answers /
if you do not know, just respond as 'I do not know' /

show only the answer alone /
Example:

Can I use the Add-on cards simultaneously?
Yes/No

Can the value on the Card be topped up from any authorised Branch?
Yes/No

Based on the above FAQ content, answer any questions the user might ask about the State Bank Vishwa Yatra Foreign Travel Card.
The answer should be based on the information provided in the FAQs.


Context: {context}
Questions: {question}


"""
```

```python
from langchain import PromptTemplate
from langchain.chains import RetrievalQA
```

```python
def sqa_prompt():
  prompt_temp = PromptTemplate(template= prompt, input_variables = ['context', 'question'])
  return prompt_temp
```

```python
def ret_answer(llm_model_name, prompt_name, v_db):
  ques_ans = RetrievalQA.from_chain_type(
      llm = llm_model_name,
      chain_type = "stuff",
      chain_type_kwargs = {"prompt": prompt_name},
      retriever = v_db.as_retriever(search_kwargs = {'k':2})
  )
  return ques_ans
```

```python
def setup_mode():
  embb = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2",
                                  model_kwargs = {"device": "cpu"})
  vdb = FAISS.load_local("/content/db/faiss", embb, allow_dangerous_deserialization= True)

  pmt = sqa_prompt()

  get_answer = ret_answer(llm_model, pmt, vdb)
  return get_answer
```

```python
llm_qa_model = setup_mode()
```
